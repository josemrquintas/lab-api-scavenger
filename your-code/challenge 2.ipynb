```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
from datetime import datetime, timedelta
```

```python
response = requests.get("https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/commits")
ds = pd.DataFrame(response.json())
ds
```

```python
norm_json = json_normalize(response.json())
ds_normalized = pd.DataFrame(norm_json)
ds_normalized.head(10)
```

```python
commit_data = json_normalize(ds['commit'])
commit_data.head()
```

```python
commit_data["date_clean"] = pd.to_datetime(commit_data["author.date"])
```
